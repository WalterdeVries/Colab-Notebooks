```python
# prompt: I need a program to produce the properties of CO2 at a given temperature and pressure, using the coolprop library

!pip install CoolProp

import CoolProp.CoolProp as CP

# Input temperature and pressure
temperature = 298.15  # Temperature in Kelvin
pressure = 101325  # Pressure in Pascals

# Define the fluid
fluid = "CO2"

# Calculate properties using CoolProp
density = CP.PropsSI("D", "T", temperature, "P", pressure, fluid)
enthalpy = CP.PropsSI("H", "T", temperature, "P", pressure, fluid)
entropy = CP.PropsSI("S", "T", temperature, "P", pressure, fluid)
specific_heat_cp = CP.PropsSI("C", "T", temperature, "P", pressure, fluid)

# Print the results
print(f"Properties of CO2 at T = {temperature} K and P = {pressure} Pa:")
print(f"Density: {density:.2f} kg/m^3")
print(f"Enthalpy: {enthalpy:.2f} J/kg")
print(f"Entropy: {entropy:.2f} J/(kg·K)")
print(f"Specific heat at constant pressure (Cp): {specific_heat_cp:.2f} J/(kg·K)")
```
